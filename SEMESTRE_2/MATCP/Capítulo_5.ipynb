## EXERCÍCIO 13

```python
from scipy import stats
import numpy as np

# 1.1) Cálculo da média amostral e do desvio padrão amostral

# Intervalos de peso e frequências
intervalos = [(53.5, 58.5), (58.5, 63.5), (63.5, 68.5),
(68.5, 73.5), (73.5, 78.5), (78.5, 83.5), (83.5, 88.5)]

frequencias = [3, 5, 9, 12, 5, 4, 2]

# Calculando as marcas de classe
marcas_classe = [(a + b) / 2 for a, b in intervalos]

# Média amostral
media_amostral = sum(f * x for f, x in zip(frequencias, marcas_classe)) / sum(frequencias)

# Desvio padrão amostral
desvio_padrao_amostral = np.sqrt(
    sum(f * (x - media_amostral)**2 for f, x in zip(frequencias, marcas_classe)) / (sum(frequencias) - 1)
)

print(f'1.1: A média amostral é {media_amostral:.2f} kg.')
print(f'    O desvio padrão amostral é {desvio_padrao_amostral:.2f} kg.')

# 1.2) Teste de hipótese unilateral
# H0: mu = 67 kg
# H1: mu > 67 kg
# Nível de significância alpha = 0.05

mu0 = 67            # Média populacional sob H0
n = sum(frequencias)  # Tamanho da amostra
s = desvio_padrao_amostral  # Desvio padrão amostral
x_barra = media_amostral  # Média amostral

# Estatística de teste t
t_calculado = (x_barra - mu0) / (s / np.sqrt(n))

# Valor crítico t unilateral (alfa = 0.05, gl = n - 1)
gl = n - 1
valor_critico = stats.t.ppf(1 - 0.05, gl)

# Decisão
rejeita_H0 = t_calculado > valor_critico

print(f'\n1.2: Estatística de teste t = {t_calculado:.2f}')
print(f'     Valor crítico t = {valor_critico:.2f}')
if rejeita_H0:
    print('     Conclusão: Rejeita-se H0. Há evidências de que o peso médio aumentou.')
else:
    print('     Conclusão: Não se rejeita H0. Não há evidências suficientes de aumento no peso médio.')

# Como xbarra pertence á regiao de rejeiçao, entao, á evidencia estatistica com nivel de significancia de 5%, para desconfiar de um aumento do peso médio dos novos estudantes
```

```python
from scipy.stats import norm

# Dados
p_hat = 29 / 32  # Proporção observada
p_0 = 0.8        # Proporção esperada
n = 32           # Tamanho da amostra
alpha = 0.05     # Nível de significância

# Estatística de teste
z_calculado = (p_hat - p_0) / np.sqrt((p_0 * (1 - p_0)) / n)
print(f"Estátistica de teste Z = {z_calculado:.3f}")

# Valor crítico para alpha = 0.05
z_critico = norm.ppf(1 - alpha)
print(f"Valor crítico Z = {z_critico:.3f}")

# Decisão
if z_calculado > z_critico:
    print("Rejeita-se H0: O novo método melhora o aproveitamento.")
else:
    print("Não se rejeita H0: Não há evidências de melhoria.")

# Cálculo do número mínimo de alunos na amostra
z_critico_contrario = 1.645  # Para rejeitar H0
p_hat = 0.90625
p_0 = 0.8

# Rearranjo da fórmula para n
n_minimo = ((z_critico_contrario * np.sqrt(p_0 * (1 - p_0))) / (p_hat - p_0))**2
print(f"Número mínimo de alunos na amostra: {int(np.ceil(n_minimo))}")

# Cálculo do nível de significância necessário
alpha_necessario = 1 - norm.cdf(z_calculado)
print(f"Nível de significância necessário: {alpha_necessario:.4f}")
```
