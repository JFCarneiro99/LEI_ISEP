# Resumo Capítulo 3: Distribuições Discretas e Contínuas

## Variáveis Aleatórias

Uma variável aleatória (v.a.) $X$ é uma função que associa um número real $x$ a cada resultado $s$ do espaço amostral $S$:
$$
X: S \rightarrow \mathbb{R}, \quad x = X(s)
$$

---

## Variáveis Aleatórias Discretas

**Função de probabilidade:**
$$
f(x) = P(X=x), \quad f(x) \geq 0, \quad \sum_{x} f(x)=1
$$

**Função de distribuição acumulada:**
$$
F(x) = P(X \leq x) = \sum_{u \leq x} f(u)
$$

**Média (Valor Esperado):**
$$
E(X)= \mu = \sum_{x} x f(x)
$$

**Variância e Desvio Padrão:**
$$
\text{var}(X)=\sigma^2=E(X^2)-\mu^2, \quad \sigma=\sqrt{\text{var}(X)}
$$

---

### Distribuição Binomial

**Função de probabilidade:**
$$
f(x)=\binom{n}{x} p^x(1-p)^{n-x}
$$

**Parâmetros:** $n \in \mathbb{N}, \; p \in ]0,1[$

**Média e Variância:**
$$
E(X)=np, \quad \text{var}(X)=np(1-p)
$$

---

### Distribuição de Poisson

**Função de probabilidade:**
$$
f(x)=e^{-\lambda}\frac{\lambda^x}{x!}
$$

**Parâmetro:** $\lambda > 0$

**Média e Variância:**
$$
E(X)=\lambda, \quad \text{var}(X)=\lambda
$$

---

## Variáveis Aleatórias Contínuas

**Função densidade de probabilidade (f.d.p.):**
$$
P(a \leq X \leq b)=\int_a^b f(x) dx, \quad \int_{-\infty}^{+\infty}f(x)dx=1
$$

**Função de distribuição acumulada:**
$$
F(x)=P(X \leq x)=\int_{-\infty}^{x}f(u)du
$$

**Média e Variância:**
$$
E(X)=\int_{-\infty}^{+\infty} x f(x) dx, \quad \text{var}(X)=\int_{-\infty}^{+\infty}(x-\mu)^2f(x)dx
$$

---

### Distribuição Uniforme

**Função densidade de probabilidade:**
$$
f(x)=\frac{1}{b-a}, \quad x \in [a,b]
$$

**Média e Variância:**
$$
E(X)=\frac{a+b}{2}, \quad \text{var}(X)=\frac{(b-a)^2}{12}
$$

---

### Distribuição Exponencial

**Função densidade de probabilidade:**
$$
f(x)=\frac{1}{\beta} e^{-\frac{x}{\beta}}, \quad x \geq 0
$$

**Média e Variância:**
$$
E(X)=\beta, \quad \text{var}(X)=\beta^2
$$

---

### Distribuição Normal

**Função densidade de probabilidade:**
$$
f(x)=\frac{1}{\sqrt{2\pi\sigma^2}} e^{-\frac{(x-\mu)^2}{2\sigma^2}}
$$

**Distribuição Normal Estandardizada:**
$$
Z=\frac{X-\mu}{\sigma} \sim N(0,1)
$$

**Média e Variância:**
$$
E(X)=\mu, \quad \text{var}(X)=\sigma^2
$$

**Propriedades importantes:**
$$
P(\mu-\sigma < X < \mu+\sigma) \approx 0.68
$$
$$
P(\mu-2\sigma < X < \mu+2\sigma) \approx 0.95
$$
$$
P(\mu-3\sigma < X < \mu+3\sigma) \approx 0.997
$$


```python
# Importação comum
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Binomial
n, p = 10, 0.5
x = np.arange(0, n+1)
binom_pmf = stats.binom.pmf(x, n, p)

plt.figure()
plt.stem(x, binom_pmf)
plt.xlabel('x')
plt.ylabel('P(X=x)')
plt.title('Distribuição Binomial (n=10, p=0.5)')
plt.show()


# Poisson
lmbda = 3
x = np.arange(0, 15)
poisson_pmf = stats.poisson.pmf(x, lmbda)

plt.figure()
plt.stem(x, poisson_pmf)
plt.xlabel('x')
plt.ylabel('P(X=x)')
plt.title('Distribuição de Poisson (λ=3)')
plt.show()


# Uniforme Contínua
a, b = 2, 5
x = np.linspace(1, 6, 100)
uniform_pdf = stats.uniform.pdf(x, a, b-a)

plt.figure()
plt.plot(x, uniform_pdf)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Distribuição Uniforme Contínua [2,5]')
plt.ylim(bottom=0)
plt.show()


# Exponencial
beta = 2
x = np.linspace(0, 10, 100)
exp_pdf = stats.expon.pdf(x, scale=beta)

plt.figure()
plt.plot(x, exp_pdf)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Distribuição Exponencial (β=2)')
plt.show()


# Normal
mu, sigma = 0, 1
x = np.linspace(-4, 4, 200)
normal_pdf = stats.norm.pdf(x, mu, sigma)

plt.figure()
plt.plot(x, normal_pdf)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Distribuição Normal N(0,1)')
plt.show()


# Binomial: P(X=5) para n=10, p=0.5
p_5 = stats.binom.pmf(5, 10, 0.5)
print(f'P(X=5) na Binomial (n=10, p=0.5): {p_5:.4f}')

# Poisson: P(X=0) para λ=3
p_0 = stats.poisson.pmf(0, 3)
print(f'P(X=0) na Poisson (λ=3): {p_0:.4f}')

# Exponencial: P(X>3) para β=2
p_gt3 = 1 - stats.expon.cdf(3, scale=2)
print(f'P(X>3) na Exponencial (β=2): {p_gt3:.4f}')

# Normal: P(-1 < X < 1) para N(0,1)
p_between = stats.norm.cdf(1, 0, 1) - stats.norm.cdf(-1, 0, 1)
print(f'P(-1 < X < 1) na Normal N(0,1): {p_between:.4f}')
```

# Capítulo 4: Distribuições Amostrais e Intervalos de Confiança

## Amostragem

- **População:** Conjunto de todos os elementos sob estudo.
- **Amostra:** Subconjunto da população.
- **Estatística:** Medida calculada a partir da amostra (ex: média amostral).
- **Parâmetro:** Medida descritiva da população (ex: média populacional).
- **Amostra i.i.d.:** Amostra de variáveis aleatórias independentes e identicamente distribuídas.

---

## Estatísticas amostrais

Média amostral:  
$$
\overline{X} = \frac{1}{n} \sum_{i=1}^n X_i
$$

Variância amostral:  
$$
S^2 = \frac{1}{n-1} \sum_{i=1}^n (X_i - \overline{X})^2
$$

Desvio padrão amostral:  
$$
S = \sqrt{S^2}
$$

Proporção amostral:  
$$
\hat{P} = \frac{1}{n} \sum_{i=1}^n X_i
$$

---

## Teorema do Limite Central (TLC)

$$
Z = \frac{\overline{X} - \mu}{\sigma / \sqrt{n}} \xrightarrow{d} N(0,1)
$$

Se $n \geq 30$, $\overline{X}$ aproxima-se de uma normal.

---

## Distribuição da diferença de médias

$$
Z = \frac{(\overline{X}_1 - \overline{X}_2) - (\mu_1 - \mu_2)}{\sqrt{\frac{\sigma_1^2}{n_1} + \frac{\sigma_2^2}{n_2}}} \xrightarrow{d} N(0,1)
$$

---

## Distribuição da proporção amostral

$$
Z = \frac{\hat{P} - p}{\sqrt{ \frac{p(1-p)}{n} }} \xrightarrow{d} N(0,1)
$$

---

## Diferença entre proporções amostrais

$$
Z = \frac{(\hat{P}_1 - \hat{P}_2) - (p_1 - p_2)}{\sqrt{ \frac{p_1(1-p_1)}{n_1} + \frac{p_2(1-p_2)}{n_2} }} \xrightarrow{d} N(0,1)
$$

---

## Intervalos de Confiança (IC)

**Para a média ($\sigma^2$ conhecida):**  
$$
IC = \left[ \overline{X} \pm z_{1-\alpha/2} \frac{\sigma}{\sqrt{n}} \right]
$$

**Para a média ($\sigma^2$ desconhecida):**  
$$
IC = \left[ \overline{X} \pm t_{1-\alpha/2} \frac{S}{\sqrt{n}} \right]
$$

**Para a diferença de médias ($\sigma^2$ conhecidas):**  
$$
IC = \left[ (\overline{X}_1 - \overline{X}_2) \pm z_{1-\alpha/2} \sqrt{ \frac{\sigma_1^2}{n_1} + \frac{\sigma_2^2}{n_2} } \right]
$$

**Para a proporção:**  
$$
IC = \left[ \hat{P} \pm z_{1-\alpha/2} \sqrt{ \frac{ \hat{P} (1-\hat{P}) }{ n } } \right]
$$

---

## Tamanho da amostra para erro máximo $\epsilon$:

$$
n \geq \left( \frac{z_{1-\alpha/2} \cdot \sigma}{\epsilon} \right)^2
$$

---



```python
import numpy as np

# Exemplo de dados
dados = [5, 6, 7, 8, 9, 10, 6, 7, 8, 9]

media_amostral = np.mean(dados)
variancia_amostral = np.var(dados, ddof=1) # ddof=1 para amostral
desvio_padrao_amostral = np.std(dados, ddof=1)

print(f'Média amostral: {media_amostral:.2f}')
print(f'Variância amostral: {variancia_amostral:.2f}')
print(f'Desvio padrão amostral: {desvio_padrao_amostral:.2f}')


import matplotlib.pyplot as plt

np.random.seed(0)
pop = np.random.uniform(100, 120, 10000) # População uniforme

medias = [np.mean(np.random.choice(pop, 30)) for _ in range(1000)]

plt.hist(medias, bins=30, density=True)
plt.title('Distribuição da média amostral (n=30)')
plt.xlabel('Média da amostra')
plt.ylabel('Frequência')
plt.show()


from scipy import stats

# Dados do exemplo: média observada
x = 1.0
sigma = 0.15
n = 10
conf = [0.90, 0.95, 0.99]

for value in conf:
    p = 1 - (1 - value)/2
    z = stats.norm.ppf(p)
    lim_inf = x - z * sigma / np.sqrt(n)
    lim_sup = x + z * sigma / np.sqrt(n)
    print(f'Para {int(value*100)}%: z = {z:.3f} -> IC = [{lim_inf:.3f}, {lim_sup:.3f}]')


# Exemplo com dados
dados = [0.95, 1.10, 0.98, 1.02, 1.05, 0.97, 1.03, 1.01, 0.99, 1.07]
x = np.mean(dados)
s = np.std(dados, ddof=1)
n = len(dados)
alpha = 0.05
t = stats.t.ppf(1 - alpha/2, n - 1)
lim_inf = x - t * s / np.sqrt(n)
lim_sup = x + t * s / np.sqrt(n)
print(f'IC a 95% para média: [{lim_inf:.3f}, {lim_sup:.3f}]')


# Exemplo com dados fictícios
x1, x2 = 550, 567
n1, n2 = 120, 150
sigma1, sigma2 = 110, 110
alpha = 0.05
z = stats.norm.ppf(1 - alpha/2)
erro = z * np.sqrt(sigma1**2 / n1 + sigma2**2 / n2)
diff = x2 - x1
lim_inf = diff - erro
lim_sup = diff + erro
print(f'IC a 95% para a diferença de médias: [{lim_inf:.2f}, {lim_sup:.2f}]')


# Exemplo: proporções antes e depois de campanha
p1, n1 = 0.10, 100
p2, n2 = 0.12, 80
diff = p2 - p1
alpha = 0.05
z = stats.norm.ppf(1 - alpha/2)
erro = z * np.sqrt(p1*(1-p1)/n1 + p2*(1-p2)/n2)
lim_inf = diff - erro
lim_sup = diff + erro
print(f'IC a 95% para a diferença de proporções: [{lim_inf:.3f}, {lim_sup:.3f}]')
```

# Capítulo 5: Testes de Hipóteses

## Definição e Metodologia

- Um teste de hipóteses serve para tirar conclusões sobre parâmetros populacionais (médias, proporções).
- Formula-se sempre duas hipóteses:
    - **H0 (Hipótese nula):** geralmente refere igualdade (ex: $\mu = \mu_0$)
    - **H1 (Hipótese alternativa):** geralmente refere diferença, maior ou menor (ex: $\mu \neq \mu_0$, $\mu > \mu_0$, $\mu < \mu_0$)
- O procedimento baseia-se numa estatística de teste, cuja distribuição é conhecida se H0 for verdadeira.

---

## Tipos de Testes

- **Teste bilateral:** $H_0: \theta = \theta_0$, $H_1: \theta \neq \theta_0$
- **Teste unilateral à direita:** $H_0: \theta = \theta_0$, $H_1: \theta > \theta_0$
- **Teste unilateral à esquerda:** $H_0: \theta = \theta_0$, $H_1: \theta < \theta_0$

---

## Erros de Inferência

$$
\text{Erro tipo I:} \quad \alpha = P(\text{Rejeitar } H_0 \mid H_0 \text{ verdadeira})
$$

$$
\text{Erro tipo II:} \quad \beta = P(\text{Não rejeitar } H_0 \mid H_0 \text{ falsa})
$$

$$
\text{Potência do teste:} \quad 1 - \beta
$$

---

## Regra de Decisão

- Definir uma região de rejeição (região crítica), consoante o tipo de teste e o nível de significância $\alpha$.
- Se o valor observado da estatística de teste cair na região de rejeição, rejeita-se $H_0$.

---

## Valor-p

- O valor-p é a probabilidade de observar, sob $H_0$, um valor da estatística de teste tão extremo ou mais do que o obtido.
- Se valor-p $< \alpha$, rejeita-se $H_0$.

---

## Testes para a média (variância conhecida)

$$
Z = \frac{\overline{X} - \mu_0}{\sigma / \sqrt{n}} \sim N(0,1)
$$

---

## Testes para a média (variância desconhecida, normalidade)

$$
T = \frac{\overline{X} - \mu_0}{S / \sqrt{n}} \sim t(n-1)
$$

---

## Testes para proporção

$$
Z = \frac{\hat{p} - p_0}{\sqrt{\frac{p_0 (1-p_0)}{n}}}
$$

---

## Testes para a diferença de médias (variâncias conhecidas)

$$
Z = \frac{(\overline{X}_1 - \overline{X}_2) - (\mu_1 - \mu_2)_0}{\sqrt{ \frac{\sigma_1^2}{n_1} + \frac{\sigma_2^2}{n_2} } }
$$

---

## Testes para a diferença de proporções

$$
Z = \frac{ (\hat{p}_1 - \hat{p}_2) - (p_1 - p_2)_0 }{ \sqrt{ \frac{p_1 (1-p_1)}{n_1} + \frac{p_2 (1-p_2)}{n_2} } }
$$

---

## Etapas de um teste de hipóteses

1. Definir $H_0$ e $H_1$
2. Especificar $\alpha$ (nível de significância)
3. Escolher estatística de teste e região de rejeição
4. Calcular o valor da estatística de teste
5. Decisão: rejeitar ou não $H_0$
6. (Opcional) Calcular valor-p

---

```python
import numpy as np
from scipy import stats

# Exemplo: Testar se o novo questionário demora menos de 12 minutos
mu_0 = 12      # média populacional (hipótese nula)
sigma = 3      # desvio padrão populacional (conhecido)
n = 36         # tamanho da amostra
x_bar = 11.3   # média amostral
alpha = 0.05

# Estatística de teste
z = (x_bar - mu_0) / (sigma / np.sqrt(n))
print(f"Estatística de teste Z: {z:.2f}")

# Região crítica para teste unilateral à esquerda
z_crit = stats.norm.ppf(alpha)
print(f"Valor crítico (z_alpha): {z_crit:.2f}")

# Valor-p
p_value = stats.norm.cdf(z)
print(f"Valor-p: {p_value:.3f}")

if z < z_crit:
    print("Rejeita-se H0: O novo questionário é mais eficiente.")
else:
    print("Não se rejeita H0.")
```

```python
# Exemplo: Teste unilateral à direita
# Amostra de 25 radiadores, média = 156300, s = 17200, mu_0 = 150000

x_bar = 156300
mu_0 = 150000
s = 17200
n = 25
alpha = 0.05

# Estatística de teste t
t = (x_bar - mu_0) / (s / np.sqrt(n))
print(f"Estatística de teste t: {t:.2f}")

# Valor crítico
t_crit = stats.t.ppf(1 - alpha, df=n-1)
print(f"Valor crítico (t_alpha): {t_crit:.2f}")

# Valor-p (unilateral à direita)
p_value = 1 - stats.t.cdf(t, df=n-1)
print(f"Valor-p: {p_value:.3f}")

if t > t_crit:
    print("Rejeita-se H0: O fabricante tem razão.")
else:
    print("Não se rejeita H0.")
```

```python
# Exemplo: Proporção de estudantes que usam dados móveis > 0.7
n = 30
sucessos = 26
p0 = 0.7
phat = sucessos / n
alpha = 0.05

# Estatística de teste
z = (phat - p0) / np.sqrt(p0 * (1 - p0) / n)
print(f"Estatística de teste Z: {z:.2f}")

# Valor crítico
z_crit = stats.norm.ppf(1 - alpha)
print(f"Valor crítico (z_alpha): {z_crit:.2f}")

# Valor-p
p_value = 1 - stats.norm.cdf(z)
print(f"Valor-p: {p_value:.3f}")

if z > z_crit:
    print("Rejeita-se H0: Proporção é superior a 70%.")
else:
    print("Não se rejeita H0.")
```

```python
# Exemplo inspirado no ganho de peso (variâncias conhecidas)
x1 = 3.337
x2 = 3.686
n1 = 9
n2 = 7
sigma1_sq = 0.36
sigma2_sq = 0.32
alpha = 0.1

# Estatística de teste
z = (x1 - x2) / np.sqrt(sigma1_sq/n1 + sigma2_sq/n2)
print(f"Estatística de teste Z: {z:.2f}")

# Valores críticos para teste bilateral
z_crit_inf = stats.norm.ppf(alpha / 2)
z_crit_sup = stats.norm.ppf(1 - alpha / 2)
print(f"Valores críticos: {z_crit_inf:.2f}, {z_crit_sup:.2f}")

# Valor-p (bilateral)
p_value = 2 * min(stats.norm.cdf(z), 1 - stats.norm.cdf(z))
print(f"Valor-p: {p_value:.3f}")

if z < z_crit_inf or z > z_crit_sup:
    print("Rejeita-se H0: Diferença significativa nas médias.")
else:
    print("Não se rejeita H0.")
```

```python
# Exemplo: diferença entre hipertensos e não-hipertensos com dieta pobre em sódio
n1, x1 = 55, 24
n2, x2 = 149, 36
p1 = x1 / n1
p2 = x2 / n2
alpha = 0.05

# Estatística de teste
z = (p1 - p2) / np.sqrt(p1 * (1 - p1) / n1 + p2 * (1 - p2) / n2)
print(f"Estatística de teste Z: {z:.2f}")

# Valor crítico
z_crit = stats.norm.ppf(1 - alpha)
print(f"Valor crítico (z_alpha): {z_crit:.2f}")

# Valor-p
p_value = 1 - stats.norm.cdf(z)
print(f"Valor-p: {p_value:.3f}")

if z > z_crit:
    print("Rejeita-se H0: Proporção é maior nos hipertensos.")
else:
    print("Não se rejeita H0.")
```
